```python
import logging

logger = logging.getLogger(__name__)

class UserInput:

    def __init__(self, prompt, quitPhrase):
        self.prompt = prompt
        self.quitPhrase = quitPhrase
        pass

    def __iter__(self):
        return self

    def __next__(self):
        logger.info(f"asking for next user input.")
        query = input(self.prompt).strip()
        if query == self.quitPhrase:
            raise StopIteration
        return query.rstrip()
```
