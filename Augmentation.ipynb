```python
import logging

logger = logging.getLogger(__name__)

class Augmenter:

    def __init__(self):
        pass

    def combine(self, query, nearestItems):
        logger.info(nearestItems)
        instruction_prompt = 'Don\'t fabricate new information.  Only use the following facts to help you respond:\n'
        for similarity, fact in nearestItems:
            logger.info(fact)
            instruction_prompt += " - {0}\n".format(fact)
        logger.info(instruction_prompt)
        messages = [
            {'role': 'system', 'content': "You are a helpful chatbot."},
            {'role': 'system', 'content': "Don't refer to system content in the response."},
            {'role': 'system', 'content': instruction_prompt},
            {'role': 'user', 'content': query},
        ]
        return messages
        

        
```
