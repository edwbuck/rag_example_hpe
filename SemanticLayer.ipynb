```python
import logging
import ollama

logger = logging.getLogger(__name__)
    
class SemanticLayer:

    def __init__(self, model):
        self.model = model
        logger.info(f"pulling model {self.model}")
        old_status = ''
        for progress in ollama.pull(self.model, stream=True):
            status = progress.get('status')
            if old_status != status:
                logger.info(f"  {status}")
            old_status = status
        logger.info(f"pull of model {self.model} complete")

    def getEncoding(self, text):
        logger.info(f"requesting encoding from ollama model {self.model}.")
        return ollama.embed(model=self.model, input=text).get('embeddings')[0]
```
