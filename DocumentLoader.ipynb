```python
import logging

logger = logging.getLogger(__name__)

class DocumentLoader:

    def __init__(self, filename):
        self.filename = filename
        self.file = open(self.filename, "r");
        logger.info(f"opening file {self.filename} for reading.")
        self.nextLine = self.file.readline()
        self.currentLine = None
        self.lineCounter = 0

    def __iter__(self):
        return self

    def __next__(self):
        if not self.nextLine:
            logger.info(f"no more facts in {self.filename}.")
            raise StopIteration
        self.currentLine = self.nextLine
        self.nextLine = self.file.readline()
        self.lineCounter += 1
        logger.debug(f"reading fact {self.lineCounter} from {self.filename}.")
        return self.currentLine.rstrip()
        
    def close(self):
        logger.info(f"closing file {self.filename}.")
        self.file.close()
        
```
